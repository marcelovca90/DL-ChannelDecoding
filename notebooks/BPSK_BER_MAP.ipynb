```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
%matplotlib inline
```

```python
# data obtained with https://plotdigitizer.com/app
df_polar = pd.read_csv('map/polar/digitized.csv')
df_random = pd.read_csv('map/random/digitized.csv')

fig = plt.figure(figsize=(11.69,8.27))
plt.semilogy(df_polar['es_no_db'],df_polar['ber'],label='Polar code')
plt.semilogy(df_random['es_no_db'],df_random['ber'],label='Random code')
plt.grid(True, which="both")
plt.xlabel('$E_b/N_0$ [dB]')
plt.ylabel('BER')
plt.xlim(0, 10)
plt.ylim(1e-5, 1e0)
plt.title('Simulated BER of the BPSK modulation (MAP)')
plt.legend()
plt.show()
```
